5.1 核心概念通俗解释
强化学习（Reinforcement Learning, RL）是一种通过试错来学习最优决策的机器学习方法。想象一下训练一只小狗：

智能体（Agent）：小狗（我们的AI控制器）

环境（Environment）：世界（卫星和物理规律）

状态（State）：小狗看到的情景（卫星当前角度、角速度）

动作（Action）：小狗做出的行为（控制力矩）

奖励（Reward）：主人给的小零食（根据姿态好坏给出的分数）

策略（Policy）：小狗的大脑（从状态到动作的映射）

目标：学会一个策略，使得长期累积奖励最大化（让卫星又快又稳地指向目标）

在数学上，强化学习通常建模为马尔可夫决策过程（MDP）。

```python
import gymnasium as gym
import numpy as np
from stable_baselines3 import PPO

# 创建环境
env = gym.make("CartPole-v1")

# 查看状态空间和动作空间
print("观测空间:", env.observation_space)  # Box(4,)
print("动作空间:", env.action_space)       # Discrete(2)

# 随机策略测试
obs, info = env.reset()
total_reward = 0
for _ in range(200):
    action = env.action_space.sample()      # 随机动作
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    if terminated or truncated:
        break
print("随机策略总奖励:", total_reward)
```
